```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV y crea un DataFrame
df = pd.read_csv('dengue.csv', encoding='latin1', delimiter=';')
```

```python
# Agrupa el DataFrame por la columna 'departamento_residencia'
grupos_por_departamento = df.groupby('departamento_residencia')
# Ahora puedes realizar operaciones en cada grupo, por ejemplo, calcular la suma de 'cantidad' por departamento
cantidad_por_departamento = grupos_por_departamento['cantidad'].sum()
# Imprime la cantidad total por departamento
print(cantidad_por_departamento)
```

```python
# Agrupa el DataFrame por la columna 'provincia_residencia'
grupos_por_provincia = df.groupby('provincia_residencia')

# Imprime los departamentos por provincia
for provincia, datos_provincia in grupos_por_provincia:
    print(f'Provincia: {provincia}')
    print(datos_provincia['departamento_residencia'].unique())
    print('\n')
```

```python
# Imprime los departamentos por provincia
for provincia, datos_provincia in grupos_por_provincia:
    print(f'Provincia: {provincia}')
    print(datos_provincia['departamento_residencia'].unique())
    print('\n')
```

```python
# Agrupa el DataFrame por la columna 'provincia_residencia' y suma la columna 'cantidad' para cada provincia
cantidad_por_provincia = df.groupby('provincia_residencia')['cantidad'].sum()

# Imprime la cantidad total de casos por provincia
print(cantidad_por_provincia)
```

```python
# Crea un gráfico de tipo torta
cantidad_por_provincia.plot(kind='pie', autopct='%1.1f%%', startangle=90)

# Añade un título al gráfico
plt.title('Cantidad de casos de dengue por provincia')

# Ajusta el diseño para que se ajuste correctamente en la figura
plt.axis('equal')

# Muestra el gráfico
plt.show()
```

```python
# Crea un gráfico de barras
cantidad_por_provincia.plot(kind='bar')

# Añade un título al gráfico
plt.title('Cantidad de casos de dengue por provincia')

# Añade etiquetas a los ejes
plt.xlabel('Provincia')
plt.ylabel('Cantidad de casos')

# Muestra el gráfico
plt.show()
```

```python
# Pivotar los datos para obtener la cantidad de casos por provincia y grupo etario
pivot_table = df.pivot_table(index='provincia_residencia', columns='grupo_etario', values='cantidad', aggfunc='sum')

# Crear un gráfico de barras apiladas
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 6))

# Añadir título y etiquetas
plt.title('Cantidad de casos de dengue por provincia y grupo etario')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de casos')
plt.legend(title='Grupo Etario', bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Lee el archivo CSV y crea un DataFrame
df = pd.read_csv('dengue.csv', encoding='latin1', delimiter=';')

# Pivotar los datos para obtener la cantidad de casos por provincia y grupo etario
pivot_table = df.pivot_table(index='provincia_residencia', columns='grupo_etario', values='cantidad', aggfunc='sum').reset_index()

# Crear un gráfico de barras
fig = px.bar(pivot_table, x='provincia_residencia', y=pivot_table.columns[1:], title='Cantidad de casos de dengue por provincia y grupo etario', barmode='stack')

# Guardar el gráfico en un archivo HTML
fig.write_html("grafico_dengue.html")
```
